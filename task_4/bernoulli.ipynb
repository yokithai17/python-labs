```python
arr = np.arange(0, 1, 0.1)

for p in arr:
    grid = np.linspace(-1.0, 2.0, size)
    sample = sps.bernoulli.rvs(p=0.45, loc=0, size=size)

    plt.plot(
        grid,
        sps.bernoulli.cdf(grid, p=p),
        color = 'red',
        linewidth = 3,
        label = "Плотность случайной величины, p={:.2f}".format(p)
    )

    plt.legend()
    plt.grid(ls=':')
    plt.show()
```

Основываясь на графики функции распределения для бернуливской случайной велечены, она приннимает 0 значение до x=(1 - p), после чего поднимается по прямой до 1, где x = p, и дальше принимает 1.
