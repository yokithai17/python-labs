```python
import numpy as np
import scipy.stats as sps

import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets

import typing

%matplotlib inline
```

```python
size = 500
grid = np.linspace(-2, 10.0, size)
low = 2
high = 5
sample = sps.randint.rvs(low=low, high=high, size=size)

plt.scatter(
    sample,
    np.zeros(size),
    alpha = 0.4,
    label = "Случайная величина"
)

plt.hist(
    sample,
    bins = 10,
    density = True,
    alpha = 0.4,
    color = "orange"
)

plt.plot(
    grid,
    sps.randint.cdf(grid, low=low, high=high, loc=0),
    color = 'red',
    linewidth = 3,
    label = "Плотность случайной величины\nlow={} high={}".format(low, high)
)

plt.legend()
plt.grid(ls=':')
plt.show()
```

```python
low_arr = np.arange(0, 6, 1)

for low in low_arr:
    size = 500
    grid = np.linspace(-2, 10.0, size)
    sample = sps.randint.rvs(low=low, high=10, size=size)
    plt.plot(
    grid,
    sps.randint.ppf(grid, low=low, high=10, loc=0),
    color = 'red',
    linewidth = 3,
    label = "Плотность случайной величины\nlow={} high={}".format(low, 10)
    )

    plt.legend()
    plt.grid(ls=':')
    plt.show()

```

Как можно увидеть чем большой low тем выше значения функции распределения, т.е уменьшается количество исходов.

```python
high_arr = np.arange(5, 10, 1)

for high in high_arr:
    size = 500
    grid = np.linspace(-2, 10.0, size)
    low = 2
    sample = sps.randint.rvs(low=low, high=high, size=size)

    plt.plot(
    grid,
    sps.randint.cdf(grid, low=low, high=high, loc=0),
    color = 'red',
    linewidth = 3,
    label = "Плотность случайной величины\nlow={} high={}".format(low, high)
    )

    plt.legend()
    plt.grid(ls=':')
    plt.show()
```

Видно, как увеличивая high функция, поднимается т.е увеличивается количество исходов.
