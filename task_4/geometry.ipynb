```python
import numpy as np
import scipy.stats as sps

import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets

import typing

%matplotlib inline
```

```python
size = 400
grid = np.linspace(-1.0, 10, size)
p = 0.3
sample = sps.geom.rvs(p=p, loc=0, size=size)

plt.scatter(
    sample,
    np.zeros(size),
    alpha = 0.4,
    label = "Случайная величина"
)

plt.hist(
    sample,
    bins = 10,
    density = True,
    alpha = 0.4,
    color = "orange"
)

plt.plot(
    grid,
    sps.geom.ppf(grid, loc=0, p=p),
    color = 'red',
    linewidth = 3,
    label = "Плотность случайной величины\np={:.2f}".format(p)
)

plt.legend()
plt.grid(ls=':')
plt.show()
```

```python
p_arr = np.arange(0.1, 1, 0.1)

for p in p_arr:
    size = 1000
    grid = np.linspace(-1.0, 10, size)

    plt.plot(
        grid,
        sps.geom.cdf(grid, loc=0, p=p),
        color = 'red',
        linewidth = 3,
        label = "Плотность случайной величины\np={:.1f}".format(p)
    )

    plt.legend()
    plt.grid(ls=':')
    plt.show()
```

Все частицы заисключением 1 живут в вероятности больше p, соотвественное уменьшая p, мы увеличиваем кол-во исходов, дающих меньше 1.
